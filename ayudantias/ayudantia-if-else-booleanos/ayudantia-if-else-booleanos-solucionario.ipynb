## 📝 Ejercicio 3 — El micro a Talagante

Diego necesita tomar el micro a Talagante, que sale puntualmente a las 7:30 de la mañana. Para registrar la hora de llegada al paradero se usa un número entero que representa cuántos minutos han pasado desde la medianoche. Por ejemplo, si llegó a las 7:20, el número es 440. Si llegó a las 7:30 exactas, el número es 450. Si llegó a las 7:31, el número es 451.

Tu programa pedirá con `input()` un dato:

1. Los minutos de llegada al paradero.
   Respuestas posibles: cualquier número entero (ejemplo: `440`).

Según lo que respondió el usuario, tu programa debe imprimir:

| Situación | Lo que imprime tu programa |
|---|---|
| Llegó a tiempo | `¡Alcanzaste el micro! Buen viaje a Talagante.` |
| Llegó tarde | `El micro ya pasó. Próxima salida en una hora.` |

No escribas texto dentro del `input()`, o sea déjalo vacío.

### ✅ Criterios de corrección

- Convierte la entrada al tipo numérico correcto con `int()` o `float()`.
- La estructura `if/else` cubre exactamente los dos casos posibles.
- El texto impreso coincide exactamente con la salida esperada (mayúsculas, puntuación, tildes y formato).

**Casos visibles:**
- Entrada: `440` | Salida esperada: `¡Alcanzaste el micro! Buen viaje a Talagante.`

**Casos ocultos:**
- Entrada: `450` | Salida esperada: `¡Alcanzaste el micro! Buen viaje a Talagante.`
- Entrada: `451` | Salida esperada: `El micro ya pasó. Próxima salida en una hora.`
- Entrada: `500` | Salida esperada: `El micro ya pasó. Próxima salida en una hora.`
- Entrada: `0` | Salida esperada: `¡Alcanzaste el micro! Buen viaje a Talagante.`

```python
# ✅ Solución esperada
minutos = int(input())
if minutos <= 450:
    print("¡Alcanzaste el micro! Buen viaje a Talagante.")
else:
    print("El micro ya pasó. Próxima salida en una hora.")
```

## 📝 Ejercicio 5 — Almuerzo en el liceo

En un liceo, la cafetería tiene una política simple: cualquier estudiante de 4to medio almuerza gratis. También almuerza gratis quien tenga un ticket especial, aunque no sea de 4to medio. Y si alguien cumple las dos condiciones a la vez, igual paga $0. Solo quien no cumpla ninguna de las dos condiciones paga el precio normal de $1.500.

Tu programa pedirá con `input()` dos datos en este orden:

1. Si el estudiante es de 4to medio.
   Respuestas posibles: `si` o `no`.

2. Si tiene un ticket especial.
   Respuestas posibles: `si` o `no`.

Según lo que respondió el usuario, tu programa debe imprimir:

| Situación | Lo que imprime tu programa |
|---|---|
| Al menos una respuesta es `si` | `Almuerzo subvencionado: $0` |
| Las dos respuestas son `no` | `Almuerzo valor normal: $1.500` |

No escribas texto dentro del `input()`, o sea déjalo vacío.

### ✅ Criterios de corrección

- Compara el texto con `"si"` (sin mayúsculas ni espacios extra).
- Usa `or` correctamente: basta con que una condición se cumpla.
- La estructura `if/else` cubre exactamente los dos casos posibles.
- El texto impreso coincide exactamente con la salida esperada (mayúsculas, puntuación, tildes y formato).

**Casos visibles:**
- Entrada: `si → no` | Salida esperada: `Almuerzo subvencionado: $0`

**Casos ocultos:**
- Entrada: `no → si` | Salida esperada: `Almuerzo subvencionado: $0`
- Entrada: `si → si` | Salida esperada: `Almuerzo subvencionado: $0`
- Entrada: `no → no` | Salida esperada: `Almuerzo valor normal: $1.500`

```python
# ✅ Solución esperada
es_4to = input() == "si"
tiene_ticket = input() == "si"
if es_4to or tiene_ticket:
    print("Almuerzo subvencionado: $0")
else:
    print("Almuerzo valor normal: $1.500")
```
