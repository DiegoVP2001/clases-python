## 2️⃣ Introducción al Contenido Nuevo

Estudia cada concepto, ejecuta los ejemplos y observa el resultado:


### 💡 Operador `and`

`and` combina dos condiciones y devuelve `True` solo si **ambas** son verdaderas. Si cualquiera es `False`, el resultado es `False`.

  | A | B | `A and B` |
  |---|---|---|
  | `True` | `True` | `True` |
  | `True` | `False` | `False` |
  | `False` | `True` | `False` |
  | `False` | `False` | `False` |

**Idea clave:** _Con `and`, TODAS las condiciones deben cumplirse._


````
tiene_suscripcion = True
contenido_disponible = True
print("¿Puede ver el video?", tiene_suscripcion and contenido_disponible)
````

### 💡 Operador `or`

`or` combina dos condiciones y devuelve `True` si al menos una es verdadera. Solo devuelve `False` si ambas son `False`.

  | A | B | `A or B` |
  |---|---|---|
  | `True` | `True` | `True` |
  | `True` | `False` | `True` |
  | `False` | `True` | `True` |
  | `False` | `False` | `False` |

**Idea clave:** _Con `or`, basta con que UNA condición se cumpla._


````
es_suscriptor = False
tiene_prueba_gratis = True
print("¿Puede acceder al contenido?", es_suscriptor or tiene_prueba_gratis)
````

### 💡 Operador `not`

`not` invierte el valor booleano de una condición. Si es `True`, lo convierte en `False`; si es `False`, lo convierte en `True`.

**Idea clave:** _`not` voltea el resultado — convierte el sí en no y viceversa._


````
esta_en_modo_avion = True
print("¿Tiene conexión?", not esta_en_modo_avion)
````

#### ⚠️ Errores típicos a evitar

| Error | Qué ocurre | Cómo corregirlo |
|---|---|---|
| Confundir `and` con `or` | El programa autoriza o bloquea en los casos equivocados | Pregúntate: ¿deben cumplirse TODAS o ALGUNA? |
| Escribir `AND`, `OR` o `NOT` en mayúscula | `SyntaxError` | Siempre en minúscula: `and`, `or`, `not` |
| Olvidar que `not` afecta solo lo que sigue inmediatamente | Puede negar solo parte de la expresión | Usa paréntesis: `not (condicion_a and condicion_b)` |
| Escribir `"sí"` con tilde al responder el `input()` | `respuesta == "si"` devuelve `False` aunque la intención fuera decir que sí | Agrega la variante con tilde: `respuesta == "si" or respuesta == "sí"` |
