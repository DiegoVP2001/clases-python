## 🧠 Contenido Nuevo

Hasta ahora has usado `and`, `or` y `not` con variables que ya tenían el valor `True` o `False`. Pero en la práctica, las condiciones pueden venir de tres formas distintas — y todas pueden combinarse en una misma expresión.

| Forma | Cómo se escribe | Resultado |
|---|---|---|
| Booleano directo | `hay_micro = True` | Ya es `True` o `False` |
| Comparación numérica | `dinero >= 900` | `True` si se cumple, `False` si no |
| Comparación con texto | `respuesta == "si"` | `True` si escribieron `"si"`, `False` si no |

Las tres formas producen `True` o `False`, y puedes combinarlas con `and`, `or` y `not` — incluso en el mismo `print()`, con paréntesis para controlar el orden de evaluación.

### Ejemplo — ¿Puede quedarse en la Vendimia?

Sebastián quiere quedarse en la Feria de la Vendimia de Isla de Maipo. Puede hacerlo si la feria sigue activa **y** (tiene plata para comprar algo **o** trae colación de la casa).

````python
# Condición 1 — booleano hardcodeado
feria_activa = True

# Condición 2 — input numérico con comparador
lucas = int(input("¿Cuántos pesos tienes? "))
tiene_plata = lucas >= 2000

# Condición 3 — respuesta si/no con comparador
respuesta_colacion = input("¿Traes colación de la casa? (si/no): ")
trae_colacion = respuesta_colacion == "si"

# Las tres condiciones combinadas con paréntesis
print("¿Puede quedarse en la Vendimia?", feria_activa and (tiene_plata or trae_colacion))
````

*Ejemplo de ejecución con $2.500 y sin colación:*

```
¿Cuántos pesos tienes? 2500
¿Traes colación de la casa? (si/no): no
¿Puede quedarse en la Vendimia? True
```

> **Nota:** Este es solo un ejemplo. Puedes combinar 3, 4 o más condiciones de distintas formas, mezclando booleanos hardcodeados, comparaciones numéricas y respuestas de texto. Los paréntesis controlan cuál parte se evalúa primero.

Diego quiere ir a Talagante después del liceo para comprar materiales. Puede ir si tiene al menos $900 en su billetera **y** (hay micro disponible **o** hay colectivo disponible).

Diego sabe perfectamente que es cierto que pasará una micro eventualmente, pero le preguntará a la gente si pasará un colectivo.

Considera que `hay_micro` ya está definido como `True` en el código. Los otros dos datos se entregan con `input()`: la cantidad de dinero que revisará en su billetera (número entero), y si hay colectivo disponible (respuesta `"si"` o `"no"`).

## Ejercicio 2 — Acceso a una transmisión en vivo

Una plataforma de streaming transmite en vivo el concierto de la Fiesta Criolla de Isla de Maipo. Un usuario puede ver la transmisión si cumple al menos una de las siguientes condiciones: tiene suscripción activa o tiene un código de acceso gratuito.

Ambos datos se piden al usuario con `input()`: si tiene suscripción activa y si tiene código de acceso gratuito. Ambas respuestas son `"si"` o `"no"`.

Un servidor de juego en línea controla quién puede entrar. Un usuario puede acceder si tiene cuenta activa y no está baneado/a del servidor.

Ambos datos se piden al usuario con `input()`: si tiene cuenta activa y si está baneado/a. Ambas respuestas son `"si"` o `"no"`.

Muestra si el usuario puede entrar a jugar y si está definitivamente bloqueado/a para jugar.

```
¿Tienes cuenta activa? (si/no): si
¿Estás baneado/a? (si/no): si
¿Puede entrar a jugar? False
¿Está bloqueado/a para jugar? True
```

````python
respuesta_cuenta = input("¿Tienes cuenta activa? (si/no): ")
tiene_cuenta_activa = respuesta_cuenta == "si"

respuesta_ban = input("¿Estás baneado/a? (si/no): ")
esta_baneado = respuesta_ban == "si"

print("¿Puede entrar a jugar?", tiene_cuenta_activa and not esta_baneado)
print("¿Está bloqueado/a para jugar?", not (tiene_cuenta_activa and not esta_baneado))
````

El saldo se pide al usuario como número entero con `input()`. Si la tienda está en mantenimiento también se consulta con `input()` como respuesta `"si"` o `"no"`.

💡 *Pista: convierte el saldo a entero con `int()` antes de compararlo.*

Muestra si el usuario puede comprar y si hay algún problema que se lo impide.

```
¿Cuál es tu saldo disponible? 8000
```
