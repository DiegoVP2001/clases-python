## 🧠 Contenido Nuevo

Hasta ahora has usado `and`, `or` y `not` con variables que pueden venir de las siguientes maneras:

| Forma | Cómo se escribe | Resultado |
|---|---|---|
| Booleano directo | `hay_micro = True` | Ya es `True` o `False` |
| Comparación numérica | `dinero >= 900` | `True` si se cumple, `False` si no |
| Comparación con texto | `respuesta == "si"` | `True` si escribieron `"si"`, `False` si no |

Lo interesante es que se pueden combinar con `and`, `or` y `not`: 
* 3 o más condiciones en el mismo `print()`,
* con paréntesis para controlar el orden de evaluación.

### Ejemplo — ¿Puede quedarse en la Vendimia?

Sebastián quiere quedarse en la Feria de la Vendimia de Isla de Maipo. Puede hacerlo si la feria sigue activa **y** (tiene plata para comprar algo **o** trae colación de la casa).

````python
# Condición 1 — booleano directo
feria_activa = True

# Condición 2 — input numérico con comparador
lucas = int(input("¿Cuántos pesos tienes? "))
tiene_plata = lucas >= 2000

# Condición 3 — respuesta si/no con comparador
respuesta_colacion = input("¿Traes colación de la casa? (si/no): ")
trae_colacion = respuesta_colacion == "si"

# Las tres condiciones combinadas con paréntesis
print("¿Puede quedarse en la Vendimia?", feria_activa and (tiene_plata or trae_colacion))
````

*Ejemplo de ejecución con $2.500 y sin colación:*

```
¿Cuántos pesos tienes? 2500
¿Traes colación de la casa? (si/no): no
¿Puede quedarse en la Vendimia? True
```

> **Nota:** Este es solo un ejemplo. Puedes combinar 3, 4 o más condiciones de distintas formas, mezclando booleanos hardcodeados, comparaciones numéricas y respuestas de texto. Los paréntesis controlan cuál parte se evalúa primero.
