## 2️⃣ Introducción al Contenido Nuevo

Estudia cada concepto, ejecuta los ejemplos y observa el resultado:


### 💡 if anidado básico

Un `if` anidado es un `if` que vive dentro del bloque de otro `if`. La condición interior solo se evalúa si la exterior ya resultó `True`.

**Idea clave:** _La segunda condición no existe para el programa si la primera falla._


````python
sala_abierta = True
es_socio = True

if sala_abierta:
    if es_socio:
        print("Puedes entrar a jugar.")
````

### 💡 else en cada nivel

Cada nivel de anidamiento puede tener su propio `else`, lo que permite cubrir todos los caminos posibles de la decisión.

**Idea clave:** _Un `if` anidado con `else` en cada nivel tiene tres o más caminos posibles — nunca uno solo._


````python
sala_abierta = True
es_socio = False

if sala_abierta:
    if es_socio:
        print("Puedes entrar a jugar.")
    else:
        print("La sala está abierta, pero necesitas ser socio.")
else:
    print("La sala está cerrada en este momento.")
````

### 💡 La sangría como jerarquía

Cada nivel de anidamiento agrega 4 espacios de sangría. El código que está "más adentro" pertenece al bloque que lo contiene.

**Idea clave:** _Un solo espacio mal puesto cambia a qué bloque pertenece una línea — Python es estricto con esto._


````python
condicion_exterior = True
condicion_interior = True

if condicion_exterior:       # nivel 1 — 0 espacios extra
    if condicion_interior:   # nivel 2 — 4 espacios
        print("Ambas OK")    # nivel 3 — 8 espacios
    else:
        print("Solo la exterior")
else:
    print("Ninguna se cumplió")
````

### 💡 El cuerpo de un if es código normal

Dentro de un `if` no solo cabe otro `if`. Puedes crear variables, hacer cálculos, concatenar texto — cualquier código Python válido.

**Idea clave:** _El `if` define *cuándo* se ejecuta el código, no *qué tipo* de código puede ser._


````python
puntaje = 850

if puntaje >= 500:
    categoria = "Nivel avanzado"
    bonus = puntaje * 2
    print("Categoría:", categoria)
    print("Bonus acumulado:", bonus)
````

#### ⚠️ Errores típicos a evitar

| Error | Qué ocurre | Cómo corregirlo |
|---|---|---|
| Segundo `if` al mismo nivel que el primero | Se evalúan como independientes, no anidados | Verificar que el `if` interior tenga 4 espacios más que el exterior |
| Olvidar el `else` del `if` exterior | Si la primera condición falla, el programa no dice nada | Agregar `else` para cubrir ese camino |
| Anidar cuando basta con `and` | Código más largo sin necesidad | Preguntar: "¿La segunda condición depende de la primera, o se pueden verificar juntas?" |
