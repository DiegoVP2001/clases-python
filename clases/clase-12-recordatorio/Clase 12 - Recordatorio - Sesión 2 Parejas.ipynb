## Ejercicio 2 — La clasificatoria

El profe Sandoval registra el rendimiento de cada estudiante en los dos últimos entrenamientos y les pone un puntaje del 1.0 al 7.0. Para poder correr representando al liceo —con polera oficial y número de pecho— el promedio de esos dos puntajes tiene que ser **4.0 o más**. Quien no llegue al mínimo puede correr igual, pero como participante libre. El programa pide los dos puntajes, calcula el promedio y entrega el veredicto.

**El programa debe:**
- Pedir **dos puntajes** de entrenamiento (escala 1.0 a 7.0, pueden tener decimales)
- Calcular el **promedio** de los dos puntajes y mostrarlo
- Si el promedio es **4.0 o más** → mostrar que clasifica e indicar cuánto le sobró sobre el mínimo
- Si el promedio es **menor a 4.0** → mostrar que no clasifica e indicar cuánto le faltó para llegar al mínimo

<details>
<summary>💡 Pista 1</summary>

Los puntajes pueden tener decimales (como 4.5) — usa `float()` para convertirlos, no `int()`.

</details>

<details>
<summary>💡 Pista 2 — cálculos dentro del if</summary>

Dentro de un bloque `if` o `else` puedes hacer cálculos que solo ocurren si esa condición se cumple:

```python
if temperatura > 30:
    grados_extra = temperatura - 30        # solo se calcula si temperatura > 30
    print("Grados de más:", grados_extra)
else:
    grados_faltantes = 30 - temperatura    # solo se calcula si temperatura <= 30
    print("Grados de menos:", grados_faltantes)
```

</details>
