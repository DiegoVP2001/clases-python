## Ejercicio 1 — La inscripción

**Concepto clave:** comparación simple (`>=`), `if / else` sin cálculos en ninguna de las dos ramas.

Ambas ramas son solo un `print` con el nombre y el resultado — no hay aritmética.

```python
nombre_corredor = input("Nombre: ")
edad_corredor = int(input("Edad: "))

if edad_corredor >= 18:
    print(nombre_corredor, "¡corres en la categoría adulto! Tienes medalla al final.")
else:
    print(nombre_corredor, "corres en la categoría júnior.")
```

## Ejercicio 2 — La clasificatoria

**Concepto clave:** aritmética antes del `if` (calcular promedio) y dentro de cada rama (calcular diferencia).

`diferencia_puntajes` se calcula distinto en cada rama: `promedio - 4.0` en el `if` y `4.0 - promedio` en el `else`. En ambos casos el resultado es positivo porque la condición garantiza el signo.

```python
puntaje_uno = float(input("Puntaje entrenamiento 1: "))
puntaje_dos = float(input("Puntaje entrenamiento 2: "))

promedio_puntajes = (puntaje_uno + puntaje_dos) / 2
print("Promedio:", promedio_puntajes)

if promedio_puntajes >= 4.0:
    diferencia_puntajes = promedio_puntajes - 4.0
    print("Resultado: Clasificas. Puedes correr con polera oficial.")
    print("Te sobró", diferencia_puntajes, "sobre el mínimo.")
else:
    diferencia_puntajes = 4.0 - promedio_puntajes
    print("Resultado: No clasificas. Corres como participante libre.")
    print("Te faltó", diferencia_puntajes, "para llegar al mínimo.")
```

## Ejercicio 3 — El combo de recuperación

**Concepto clave:** condición compuesta con `and` + cálculo dentro del `if`.

`precio_combo` es una constante en el código, no un `input` — el estudiante debe notar que no todo valor viene del usuario. El `else` es genérico porque sin `elif` no es posible distinguir cuál de las dos condiciones falló.

```python
precio_combo = 2500
tiempo_llegada = int(input("Tiempo de llegada (minutos): "))
saldo_tarjeta = int(input("Saldo en la tarjeta: "))

if tiempo_llegada <= 60 and saldo_tarjeta >= precio_combo:
    saldo_restante = saldo_tarjeta - precio_combo
    print("¡Combo asegurado! Llegaste en tiempo y con saldo.")
    print("Te quedan $", saldo_restante, "en la tarjeta.")
else:
    print("No puedes pedir el combo premium. Tiempo o saldo no cumplen los requisitos.")
```
