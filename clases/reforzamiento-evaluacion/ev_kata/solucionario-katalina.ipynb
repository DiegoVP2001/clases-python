## Ejercicio 3 — Promedio de gastos (20 pts)

| Criterio | Puntaje |
|---|---|
| `gasto_lunes`, `gasto_miercoles`, `gasto_viernes` con valores correctos | 6 pts (2 c/u) |
| `suma_gastos` suma las tres variables | 7 pts |
| `promedio_diario` divide `suma_gastos` en 3 | 7 pts |

**Resultado esperado:** `promedio_diario = 1433.33...`

**Notas de corrección:**
- El `print()` ya estaba dado — no se evalúa.
- Si suma correctamente pero divide en 2 o en otro número: 13/20.
- Si define los gastos directamente en `suma_gastos` sin variables intermedias: descontar 3 pts (no siguió el paso a paso).

```python
# Solución Ejercicio 3

# Paso 1: Define el gasto de cada día
gasto_lunes = 1500
gasto_miercoles = 2000
gasto_viernes = 800

# Paso 2: Suma los tres gastos
suma_gastos = gasto_lunes + gasto_miercoles + gasto_viernes

# Paso 3: Divide la suma en 3 para obtener el promedio
promedio_diario = suma_gastos / 3

# Paso 4: Imprime el resultado
print("Promedio diario:", promedio_diario)
```
