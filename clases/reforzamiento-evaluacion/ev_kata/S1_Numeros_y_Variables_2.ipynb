## 📖 PARTE 1 — Python como calculadora

Python puede hacer operaciones matemáticas directamente.

| Símbolo | Operación | Ejemplo | Resultado |
|---------|-----------|---------|----------|
| `+` | Suma | `5 + 3` | `8` |
| `-` | Resta | `10 - 4` | `6` |
| `*` | Multiplicación | `3 * 7` | `21` |
| `/` | División con decimales | `10 / 3` | `3.333...` |
| `//` | División entera | `10 // 3` | `3` |
| `%` | Resto de una división | `10 % 3` | `1` |
| `**` | Potencia | `2 ** 8` | `256` |

> ⚠️ **Cuidado:** En Python los decimales van con **punto**, no con coma. `3.14` ✅ — `3,14` ❌

## 📖 PARTE 2 — Las variables

Una **variable** es como una caja con nombre donde guardas un dato.

```python
apodo = "Barbero"
nota_promedio = 7.0
goles_partido = 4
```

**Reglas para nombrar variables:**
- Usa nombres que describan qué contienen → `nota_promedio` ✅, `x` ❌
- Separa las palabras con `_` → esto se llama **snake_case**
- Sin tildes, ñ, espacios ni caracteres especiales
- No uses palabras reservadas de Python (`for`, `if`, `print`, `while`...)

📓 *Copia estas reglas en tu cuaderno.*

````python
# El profe ejecuta esto en vivo
apodo = "Barbero"
nota_promedio = 7.0
goles_partido = 4

print(apodo)
print(nota_promedio)
print(goles_partido)
````

````python
# Las variables pueden cambiar de valor
nota_promedio = 6.3
print(nota_promedio)

nota_promedio = 6.9
print(nota_promedio)

# ¿Qué pasó? Anótalo en tu cuaderno.
````
