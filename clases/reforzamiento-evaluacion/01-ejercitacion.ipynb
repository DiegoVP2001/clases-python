## 🐛 ¿Qué es depurar?

**Depurar** (o *debuggear*) significa encontrar y corregir errores en un programa que no funciona bien.

Existen tres tipos de errores:

| Tipo | Qué pasa | Ejemplos |
|---|---|---|
| **Error de sintaxis** | Python no puede leer el código y se detiene con un mensaje de error | `nota promedio = 6.0` (espacio en el nombre), `temperatura = 10,5` (coma como separador decimal), `nombre = "Juan` (comilla sin cerrar) |
| **Error de lógica** | El código se ejecuta, pero da un resultado incorrecto | Dividir por 3 cuando deberías dividir por 2 |
| **Error de indentación** | Una línea tiene sangría inesperada al inicio del programa | `.....resultado = 10` al inicio (sin bloque que lo justifique) |

**Estrategia para depurar:**
1. Si hay mensaje de error, lee la línea que indica.
2. Compara el resultado que obtienes con el resultado esperado.
3. Corrige un error a la vez y vuelve a ejecutar.
