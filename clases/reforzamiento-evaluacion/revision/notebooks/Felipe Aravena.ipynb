## Ejercicio 1 — Promedio de notas ⭐ (10 pts)

## Ejercicio 5 — Repartir ganancias del puesto de dulces ⭐⭐ (15 pts)

Tú y tus compañeros pusieron un puesto de dulces en el liceo.
Al final del día:

1. Cuentan el dinero,
2. Restan lo que gastaron en insumos,
3. Y se reparten lo que sobra en partes iguales.

La **ganancia neta es la diferencia entre lo recaudado y lo gastado** en insumos.

**El programa debe pedir con `input()`:**
- El total recaudado durante el día
- El total gastado en insumos (golosinas, bolsas, etc.)
- La cantidad de personas en el puesto

A continuación un ejemplo de lo que **el programa debe imprimir con `print()`:**
```
--- REPARTO DEL PUESTO ---
Total recaudado: $ 42000.0
Gastos en insumos: $ 18000.0
Ganancia neta: $ 24000.0
Integrantes: 4
Ganancia por persona: $ 6000.0
```

````python
# Ejercicio 5 — Repartir ganancias del puesto de dulces
# Escribe tu código aquí
total_recaudado = float(input("Dinero recaudado: "))
gastos_insumos = float(input("Gastos en insumos: "))
ganancia_neta = total_recaudado - gastos_insumos
integrantes = int(input("Total de integrantes: "))
ganancia_individual = ganancia_neta/integrantes
print()
print("--- REPARTO DEL PUESTO ---")
print("Total recaudado:",total_recaudado,sep=" $ ")
print("Gastos en insumos:",gastos_insumos,sep=" $ ")
print("Ganancia neta:",ganancia_neta,sep=" $ ")
print("Integrantes:",integrantes)
print("Ganancia por persona:",ganancia_individual,sep=" $ ")
````

## Ejercicio 6 — Depurar cálculo de nota ponderada ⭐⭐ (15 pts)

El siguiente programa tiene **5 errores**. Encuéntralos y corrígelos sin cambiar los valores.

**El programa debe imprimir con `print()`:**
```
📋 CÁLCULO DE NOTA FINAL
Controles (35%): 5.8
Módulo (30%): 6.4
Proyecto (20%): 7.0
Participación (15%): 5.5
Nota final ponderada: 6.175
```

Verificación: 5.8×0.35 + 6.4×0.30 + 7.0×0.20 + 5.5×0.15 = 6.175


````python
# Ejercicio 6 — Depurar cálculo de nota ponderada
# Este código tiene 5 errores. Corrígelos sin cambiar los valores.

nota_controles = 5.8
nota_modulo = 6.4
nota_proyecto = 7.0
nota_participacion = 5.5

nota_final = (nota_controles * 0.35) + (nota_modulo * 0.30) + (nota_proyecto * 0.20) + (nota_participacion * 0.15)

print("📋 CÁLCULO DE NOTA FINAL")
print("Controles (35%):", nota_controles)
print("Módulo (30%):", nota_modulo)
print("Proyecto (20%):", nota_proyecto)
print("Participación (15%)", nota_participacion)
print("Nota final ponderada:", nota_final)
````

## Ejercicio 7 — Registro semanal de gastos ⭐⭐⭐ (20 pts)

Para ordenar los gastos, decidiste hacer un programa que registra tus gastos de lunes a viernes y el dinero que tienes al inicio de la semana.

**El programa debe pedir con `input()`:**
- El presupuesto semanal disponible
- El gasto del lunes
- El gasto del martes
- El gasto del miércoles
- El gasto del jueves
- El gasto del viernes

Luego, **se debe calcular:**
1. El gasto total
2. El gasto promedio
3. La diferencia entre presupuesto semanal y gasto total

A continuación se presenta un ejemplo de lo que **el programa debe imprimir con `print()`:**
```
--- RESUMEN SEMANAL ---
Presupuesto inicial: $ 30000
Total gastado: $ 22500
Promedio de gasto: $ 4500
Saldo restante: $ 7500
```
**Atención:** No hay problema si el saldo restante es negativo, eso dependerá de los inputs que se den.

````python
# Ejercicio 7 — Registro semanal de gastos
# Escribe tu código aquí
presupuesto_inicial = float(input("Presupuesto inicial: "))
gasto_lunes = float(input("Gasto del dia lunes: "))
gasto_martes = float(input("Gasto del dia martes: "))
gasto_miercoles = float(input("Gasto del dia miercoles: "))
gasto_jueves = float(input("Gasto del dia jueves: "))
gasto_viernes= float(input("Gasto del dia viernes: "))
gasto_total = gasto_lunes + gasto_martes + gasto_miercoles + gasto_jueves + gasto_viernes
promedio_gasto = gasto_total/5
presupuesto_restante = presupuesto_inicial - gasto_total
print()
print("---RESUMEN SEMANAL---")
print("Presupuesto inicial:",presupuesto_inicial,sep=" $ ")
print("Total gastado:",gasto_total,sep=" $ ")
print("Promedio de gasto:",promedio_gasto,sep=" $ ")
print("Saldo restante:",presupuesto_restante,sep=" $ ")
````
