## Ejercicio 5 — Repartir ganancias del puesto de dulces ⭐⭐ (15 pts)

Al final del día:

1. Cuentan el dinero,

## Ejercicio 6 — Depurar cálculo de nota ponderada ⭐⭐ (15 pts)

El siguiente programa tiene **5 errores**. Encuéntralos y corrígelos sin cambiar los valores.

**El programa debe imprimir con `print()`:**
```
📋 CÁLCULO DE NOTA FINAL
Controles (35%): 5.8
Módulo (30%): 6.4
Proyecto (20%): 7.0
Participación (15%): 5.5
Nota final ponderada: 6.175
```

Verificación: 5.8×0.35 + 6.4×0.30 + 7.0×0.20 + 5.5×0.15 = 6.175


````python
# Ejercicio 6 — Depurar cálculo de nota ponderada
# Este código tiene 5 errores. Corrígelos sin cambiar los valores.

nota_controles = 5.8
nota_modulo = 6.4
nota_proyecto = 7.0
nota_participacion = 5.5

nota_final = (nota_controles * 0.35) + (nota_modulo * 0.30) + (nota_proyecto * 0.20) + (nota_participacion * 0.15)

print("📋 CÁLCULO DE NOTA FINAL")
print("Controles (35%):", nota_controles)
print("Módulo (30%):", nota_modulo)
print("Proyecto (20%):", nota_proyecto)
print("Participación (15%)", nota_participacion)
print("Nota final ponderada:", nota_final)
````

## Ejercicio 7 — Registro semanal de gastos ⭐⭐⭐ (20 pts)

Para ordenar los gastos, decidiste hacer un programa que registra tus gastos de lunes a viernes y el dinero que tienes al inicio de la semana.

**El programa debe pedir con `input()`:**
- El presupuesto semanal disponible
- El gasto del lunes
- El gasto del martes
- El gasto del miércoles
- El gasto del jueves
- El gasto del viernes

Luego, **se debe calcular:**
1. El gasto total
2. El gasto promedio
3. La diferencia entre presupuesto semanal y gasto total

A continuación se presenta un ejemplo de lo que **el programa debe imprimir con `print()`:**
```
--- RESUMEN SEMANAL ---
Presupuesto inicial: $ 30000
Total gastado: $ 22500
Promedio de gasto: $ 4500
Saldo restante: $ 7500
```
**Atención:** No hay problema si el saldo restante es negativo, eso dependerá de los inputs que se den.

````python
# Ejercicio 7 — Registro semanal de gastos
# Escribe tu código aquí

presupuesto_inicial = float(input("presupuesto inicial:"))

print("presupuesto inicial: ", presupuesto_inicial)

gasto_del_lunes = float(input("gasto lunes:"))

gasto_del_martes = float(input("gasto martes"))

gasto_del_miercoles = float(input("gasto miercoles:"))

gasto_del_jueves = float(input("gasto jueves"))

gasto_del_viernes = float(input("gasto viernes"))

total_gastado = gasto_del_lunes + gasto_del_martes + gasto_del_miercoles + gasto_del_jueves + gasto_del_viernes

print("total gastado:", total_gastado)

promedio_de_gasto = total_gastado / 5

print("promedio de gasto", promedio_de_gasto)

saldo_restante = presupuesto_inicial - total_gastado

print("saldo restante", saldo_restante)




````
