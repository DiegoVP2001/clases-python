2. Depósito de sueldo: +$85.000

3. Pago de arriendo: −$45.000

4. Compra de supermercado: −$18.000

5. Depósito de bono: +$10.000

6. Pago de servicios básicos: −$8.500

Debes usar una sola variable que se va actualizando. Para ello, puedes usar la forma detallada (`variable = variable + valor`) o los operadores abreviados (`+=`, `-=`). Ambas son válidas.

**El programa debe imprimir con `print()`:**

````python
# Ejercicio 3 — Simulador de ahorro mensual
# Escribe tu código aquí

saldo = 20000
print(f"Saldo inicial: $ {saldo}")

saldo += 85000
print(f"Tras sueldo: $ {saldo}")

saldo -= 45000
print(f"Tras arriendo: $ {saldo}")

saldo -= 18000
print(f"Tras supermercado: $ {saldo}")

saldo += 10000
print(f"Tras bono: $ {saldo}")

saldo -= 8500
print(f"Saldo final: $ {saldo}")
````

## Ejercicio 4 — Calculadora de descuento en tienda ⭐⭐ (15 pts)

Como parte del cyberday, debes crear un **programa capaz de pedir con `input()`:**
- El precio original del producto
- El porcentaje de descuento

A continuación, se muestra un ejemplo de lo que **programa debe imprimir con `print()`** si alguien ingresa un producto que cuesta $59.990 y tiene un descuento del 20%.
```
--- RESUMEN DEL DESCUENTO ---
Precio original: $ 59990.0
Descuento (20%): $ 11998.0
Precio final: $ 47992.0
```


````python
# Ejercicio 4 — Calculadora de descuento en tienda
# Escribe tu código aquí

precio_original_str = input("Ingrese el precio original del producto: ")
porcentaje_descuento_str = input("Ingrese el porcentaje de descuento: ")

precio_original = float(precio_original_str)
porcentaje_descuento = float(porcentaje_descuento_str)

descuento_monto = precio_original * (porcentaje_descuento / 100)
precio_final = precio_original - descuento_monto

print("RESUMEN DEL DESCUENTO")
print(f"Precio original:$ {precio_original:.1f}")
print(f"Descuento ({int(porcentaje_descuento)}%): $ {descuento_monto:.1f}")
print(f"Precio final: $ {precio_final:.1f}")
````

## Ejercicio 5 — Repartir ganancias del puesto de dulces ⭐⭐ (15 pts)

Tú y tus compañeros pusieron un puesto de dulces en el liceo.
Al final del día:

1. Cuentan el dinero,
2. Restan lo que gastaron en insumos,
3. Y se reparten lo que sobra en partes iguales.

La **ganancia neta es la diferencia entre lo recaudado y lo gastado** en insumos.

**El programa debe pedir con `input()`:**
- El total recaudado durante el día
- El total gastado en insumos (golosinas, bolsas, etc.)
- La cantidad de personas en el puesto

A continuación un ejemplo de lo que **el programa debe imprimir con `print()`:**
```
--- REPARTO DEL PUESTO ---
Total recaudado: $ 42000.0
Gastos en insumos: $ 18000.0
Ganancia neta: $ 24000.0
Integrantes: 4
Ganancia por persona: $ 6000.0
```

````python
# Ejercicio 5 — Repartir ganancias del puesto de dulces
# Escribe tu código aquí

total_recaudado_str = input("Ingrese el total recaudado durante el día: ")
total_gastado_str = input("Ingrese el total gastado en insumos: ")
cantidad_personas_str = input("Ingrese la cantidad de personas en el puesto: ")

total_recaudado = float(total_recaudado_str)
total_gastado = float(total_gastado_str)
cantidad_personas = int(cantidad_personas_str)

ganancia_neta = total_recaudado - total_gastado
ganancia_por_persona = ganancia_neta / cantidad_personas

print("--- REPARTO DEL PUESTO ---")
print("Total recaudado: $ {total_recaudado:.1f}")
print("Gastos en insumos: $ {total_gastado:.1f}")
print("Ganancia neta: $ {ganancia_neta:.1f}")
print("Integrantes: {cantidad_personas}")
print("Ganancia por persona: $ {ganancia_por_persona:.1f}")
````

## Ejercicio 6 — Depurar cálculo de nota ponderada ⭐⭐ (15 pts)

El siguiente programa tiene **5 errores**. Encuéntralos y corrígelos sin cambiar los valores.

**El programa debe imprimir con `print()`:**
```
📋 CÁLCULO DE NOTA FINAL
Controles (35%): 5.8
Módulo (30%): 6.4
Proyecto (20%): 7.0
Participación (15%): 5.5
Nota final ponderada: 6.175
```

Verificación: 5.8×0.35 + 6.4×0.30 + 7.0×0.20 + 5.5×0.15 = 6.175


````python
# Ejercicio 6 — Depurar cálculo de nota ponderada
# Este código tiene 5 errores. Corrígelos sin cambiar los valores.

nota_controles = 5.8
nota_modulo = 6.4
nota_proyecto = 7.0
nota_participacion = 5.5

nota_final = (nota_controles * 0.35) + (nota_modulo * 0.30) + (nota_proyecto * 0.20) + (nota_participacion * 0.15)

print("📋 CÁLCULO DE NOTA FINAL")
print("Controles (35%):", nota_controles)
print("Módulo (30%):", nota_modulo)
print("Proyecto (20%):", nota_proyecto)
print("Participación (15%)", nota_participacion)
print("Nota final ponderada:", nota_final)
````

## Ejercicio 7 — Registro semanal de gastos ⭐⭐⭐ (20 pts)

Para ordenar los gastos, decidiste hacer un programa que registra tus gastos de lunes a viernes y el dinero que tienes al inicio de la semana.

**El programa debe pedir con `input()`:**
- El presupuesto semanal disponible
- El gasto del lunes
- El gasto del martes
- El gasto del miércoles
- El gasto del jueves
- El gasto del viernes

Luego, **se debe calcular:**
1. El gasto total
2. El gasto promedio
3. La diferencia entre presupuesto semanal y gasto total

A continuación se presenta un ejemplo de lo que **el programa debe imprimir con `print()`:**
```
--- RESUMEN SEMANAL ---
Presupuesto inicial: $ 30000
Total gastado: $ 22500
Promedio de gasto: $ 4500
Saldo restante: $ 7500
```
**Atención:** No hay problema si el saldo restante es negativo, eso dependerá de los inputs que se den.

````python
# Ejercicio 7 — Registro semanal de gastos
# Escribe tu código aquí

presupuesto_semanal_str = input("Ingrese el presupuesto semanal disponible: ")
gasto_lunes_str = input("Ingrese el gasto del lunes: ")
gasto_martes_str = input("Ingrese el gasto del martes: ")
gasto_miercoles_str = input("Ingrese el gasto del miércoles: ")
gasto_jueves_str = input("Ingrese el gasto del jueves: ")
gasto_viernes_str = input("Ingrese el gasto del viernes: ")

presupuesto_semanal = float(presupuesto_semanal_str)
gasto_lunes = float(gasto_lunes_str)
gasto_martes = float(gasto_martes_str)
gasto_miercoles = float(gasto_miercoles_str)
gasto_jueves = float(gasto_jueves_str)
gasto_viernes = float(gasto_viernes_str)

gasto_total = gasto_lunes + gasto_martes + gasto_miercoles + gasto_jueves + gasto_viernes
gasto_promedio = gasto_total / 5
saldo_restante = presupuesto_semanal - gasto_total

print("--- RESUMEN SEMANAL ---")
print(f"Presupuesto inicial: $ {presupuesto_semanal:.0f}")
print(f"Total gastado: $ {gasto_total:.0f}")
print(f"Promedio de gasto: $ {gasto_promedio:.0f}")
print(f"Saldo restante: $ {saldo_restante:.0f}")
````
