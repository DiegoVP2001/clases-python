## Ejercicio 3 — Simulador de ahorro mensual ⭐⭐ (15 pts)

Escribe un programa que simule los siguientes movimientos de una cuenta de ahorro:

1. Saldo inicial: $20.000

2. Depósito de sueldo: +$85.000

3. Pago de arriendo: −$45.000

4. Compra de supermercado: −$18.000

5. Depósito de bono: +$10.000

6. Pago de servicios básicos: −$8.500

Debes usar una sola variable que se va actualizando. Para ello, puedes usar la forma detallada (`variable = variable + valor`) o los operadores abreviados (`+=`, `-=`). Ambas son válidas.

**El programa debe imprimir con `print()`:**
```
Saldo inicial: $ 20000
Tras sueldo: $ 105000
Tras arriendo: $ 60000
Tras supermercado: $ 42000
Tras bono: $ 52000
Saldo final: $ 43500
```
**Pista:** escribe un print después de cada operación.

````python
# Ejercicio 3 — Simulador de ahorro mensual
# Escribe tu código aquí
saldo=20000
sueldo=85000
arriendo=45000
compras_en_supermercado=18000
bono=10000
servicios_basicos=8500

print("Saldo inicial:", saldo, sep=" $ ")
saldo+=sueldo
print("Tras sueldo:", saldo, sep=" $ ")
saldo-=arriendo
print("Tras arriendo:", saldo, sep=" $ ")
saldo-=compras_en_supermercado
print("Tras supermercado:", saldo, sep=" $ ")
saldo+=bono
print("Tras bono:", saldo, sep=" $ ")
saldo-=servicios_basicos
print("Saldo final:", saldo, sep=" $ ")
````

## Ejercicio 6 — Depurar cálculo de nota ponderada ⭐⭐ (15 pts)

El siguiente programa tiene **5 errores**. Encuéntralos y corrígelos sin cambiar los valores.

**El programa debe imprimir con `print()`:**
```
📋 CÁLCULO DE NOTA FINAL
Controles (35%): 5.8
Módulo (30%): 6.4
Proyecto (20%): 7.0
Participación (15%): 5.5
Nota final ponderada: 6.175
```

Verificación: 5.8×0.35 + 6.4×0.30 + 7.0×0.20 + 5.5×0.15 = 6.175


````python
# Ejercicio 6 — Depurar cálculo de nota ponderada
# Este código tiene 5 errores. Corrígelos sin cambiar los valores.

nota_controles = 5.8
nota_modulo = 6.4
nota_proyecto = 7.0
nota_participacion = 5.5

nota_final = (nota_controles * 0.35) + (nota_modulo * 0.30) + (nota_proyecto * 0.20) + (nota_participacion * 0.15)

print("📋 CÁLCULO DE NOTA FINAL")
print("Controles (35%):", nota_controles)
print("Módulo (30%):", nota_modulo)
print("Proyecto (20%):", nota_proyecto)
print("Participación (15%)", nota_participacion)
print("Nota final ponderada:", nota_final)
````

## Ejercicio 7 — Registro semanal de gastos ⭐⭐⭐ (20 pts)

Para ordenar los gastos, decidiste hacer un programa que registra tus gastos de lunes a viernes y el dinero que tienes al inicio de la semana.

**El programa debe pedir con `input()`:**
- El presupuesto semanal disponible
- El gasto del lunes
- El gasto del martes
- El gasto del miércoles
- El gasto del jueves
- El gasto del viernes

Luego, **se debe calcular:**
1. El gasto total
2. El gasto promedio
3. La diferencia entre presupuesto semanal y gasto total

A continuación se presenta un ejemplo de lo que **el programa debe imprimir con `print()`:**
```
--- RESUMEN SEMANAL ---
Presupuesto inicial: $ 30000
Total gastado: $ 22500
Promedio de gasto: $ 4500
Saldo restante: $ 7500
```
**Atención:** No hay problema si el saldo restante es negativo, eso dependerá de los inputs que se den.

````python
# Ejercicio 7 — Registro semanal de gastos
# Escribe tu código aquí
presupuesto=float(input("Dime tu presupuesto inicial para esta semana: "))
gasto_lunes=float(input("Dime tu gasto del dia Lunes: "))
gasto_martes=float(input("Dime tu gasto del dia Mares: "))
gasto_miercoles=float(input("Dime tu gasto del dia Miercoles: "))
gasto_jueves=float(input("Dime tu gasto del dia Jueves: "))
gasto_viernes=float(input("Y por ultimo dime tu gasto del dia Viernes: "))

gasto_total=gasto_lunes+gasto_martes+gasto_miercoles+gasto_jueves+gasto_viernes
gasto_promedio=gasto_total/5
saldo_final=presupuesto-gasto_total
print("------RESUMEN SEMANA-------")
print("Presupuesto inicial:", presupuesto, sep=" $ ")
print("Total gastado:", gasto_total, sep=" $ ")
print("Promedio de gasto:", gasto_promedio, sep=" $ ")
print("Saldo restante:", saldo_final, sep=" $ ")
````
