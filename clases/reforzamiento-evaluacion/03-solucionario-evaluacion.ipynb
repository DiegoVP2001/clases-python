## Ejercicio 4 — Calculadora de descuento en tienda ⭐⭐ (15 pts)

```python
precio_original = float(input("¿Cuál es el precio original? "))
porcentaje_descuento = float(input("¿Qué porcentaje de descuento tiene? "))

monto_descuento = precio_original * (porcentaje_descuento / 100)
precio_final = precio_original - monto_descuento

print("--- RESUMEN DEL DESCUENTO ---")
print("Precio original: $", precio_original)
print("Descuento (" + str(int(porcentaje_descuento)) + "%): $", monto_descuento)
print("Precio final: $", precio_final)
```

**Pauta Ej. 4:**

**Aceptable:** no es necesario que el porcentaje aparezca dinámico en el print — se acepta `print("Descuento: $", monto_descuento)` también.

**Con 59990 y 20%:** descuento = 11998.0, precio final = 47992.0

## Ejercicio 5 — Repartir ganancias del puesto de dulces ⭐⭐ (15 pts)

```python
total_recaudado = float(input("¿Cuánto recaudaron en total? "))
gastos_insumos = float(input("¿Cuánto gastaron en insumos? "))
cantidad_integrantes = int(input("¿Cuántas personas hay en el puesto? "))

ganancia_neta = total_recaudado - gastos_insumos
ganancia_por_persona = ganancia_neta / cantidad_integrantes

print("--- REPARTO DEL PUESTO ---")
print("Total recaudado: $", total_recaudado)
print("Gastos en insumos: $", gastos_insumos)
print("Ganancia neta: $", ganancia_neta)
print("Integrantes:", cantidad_integrantes)
print("Ganancia por persona: $", ganancia_por_persona)
```

**Con 42000, 18000, 4 personas:** ganancia neta = 24000.0, ganancia por persona = 6000.0

## Ejercicio 6 — Depurar cálculo de nota ponderada ⭐⭐ (15 pts)

**Los 3 errores eran:**
1. `nota controles` → `nota_controles` (espacio en nombre de variable)
2. `nota_modulo + 0.30` → `nota_modulo * 0.30` (suma en lugar de multiplicación)
3. `print("Participación (15%)" nota_participacion)` → falta la coma antes de `nota_participacion`


```python
nota_controles = 5.8
nota_modulo = 6.4
nota_proyecto = 7.0
nota_participacion = 5.5

nota_final = (nota_controles * 0.35) + (nota_modulo * 0.30) + (nota_proyecto * 0.20) + (nota_participacion * 0.15)

print("📋 CÁLCULO DE NOTA FINAL")
print("Controles (35%):", nota_controles)
print("Módulo (30%):", nota_modulo)
print("Proyecto (20%):", nota_proyecto)
print("Participación (15%):", nota_participacion)
print("Nota final ponderada:", nota_final)
```

**Pauta Ej. 6:**
- 4 pts por cada error correctamente identificado y corregido (3 × 4 = 12 pts)
- 3 pts: el resultado final impreso es 6.175

## Ejercicio 7 — Registro semanal de gastos ⭐⭐⭐ (20 pts)
