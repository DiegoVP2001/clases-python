## Ejercicio 3 — Tarjeta de transporte ⭐⭐ (15 pts)

Escribe un programa que simule los siguientes movimientos de una tarjeta de transporte:

1. Saldo inicial: $12.000
2. Recarga de la mañana: +$5.000
3. Viaje al liceo: −$850
4. Compra de colación: −$1.500
5. Recarga de la tarde: +$3.000
6. Viaje de regreso: −$850

**Regla obligatoria:** debes usar **una sola variable de saldo** que se va actualizando paso a paso. Puedes usar la forma corta (`+=`, `-=`) o la forma larga (`variable = variable + valor`, `variable = variable - valor`), pero siempre debe ser **la misma variable**. No hagas una variable distinta para cada movimiento.

**El programa debe imprimir con `print()`:**
```
Saldo inicial: $ 12000
Tras recarga mañana: $ 17000
Tras viaje al liceo: $ 16150
Tras colación: $ 14650
Tras recarga tarde: $ 17650
Saldo final: $ 16800
```


````python
# Ejercicio 3 — Tarjeta de transporte
# Escribe tu código aquí

Saldo_inicial = 12000
recarga_mañana = 5000
suma1 = Saldo_inicial + recarga_mañana
print("Saldo inicial: $", Saldo_inicial)
print("Tras Recarga mañana: $", suma1)
viaje_liceo= 850
resta1= suma1 - viaje_liceo
print("Tras viaje al liceo: $", resta1)
colacion = 1500
resta2 = resta1 - colacion
print("Tras colacion: $", resta2)
recarga_tarde = 3000
suma2 = resta2 + recarga_tarde
print("Tras recarga tarde: $", suma2)
print("Saldo final: $", suma2 - viaje_liceo)
````

## Ejercicio 5 — Repartir fondos de una actividad de curso ⭐⭐ (15 pts)

Tu curso organiza una actividad y junta dinero para financiarla.
Después de pagar los materiales, quieren saber cuánto dinero queda y cuánto corresponde por persona si se reparte en partes iguales.

**El programa debe pedir con `input()`:**
- El total reunido por el curso
- El total gastado en materiales
- La cantidad de personas que participaron

**El programa debe imprimir con `print()`:**
```
--- REPARTO DE FONDOS ---
Total reunido: $ 64000.0
Gasto en materiales: $ 22000.0

```

## Ejercicio 6 — Depurar cálculo de promedio ponderado ⭐⭐ (15 pts)

El siguiente programa tiene **5 errores**. Encuéntralos y corrígelos sin cambiar los valores.

**El programa debe imprimir con `print()`:**
```
📋 PROMEDIO PONDERADO DEL TALLER
Ejercicios (40%): 6.0
Desafío (25%): 5.4
Bitácora (20%): 6.8
Participación (15%): 6.4
Promedio final: 6.07
```

Verificación: 6.0×0.40 + 5.4×0.25 + 6.8×0.20 + 6.4×0.15 = 6.07


````python
# Ejercicio 6 — Depurar cálculo de promedio ponderado
# Este código tiene 5 errores. Corrígelos sin cambiar los valores.

nota_ejercicios = 6.0
nota_desafio = 5.4
nota_bitacora = 6.8
nota_participacion = 6.4

promedio_final = (nota_ejercicios * 0.40) + (nota_desafio * 0.25) + (nota_bitacora * 0.20) + (nota_participacion * 0.15)

print("📋 PROMEDIO PONDERADO DEL TALLER")
print("Ejercicios (40%):", nota_ejercicios)
print("Desafío (25%):", nota_desafio)
print("Bitácora (20%):", nota_bitacora)
print("Participación (15%)", nota_participacion)
print("Promedio final:", promedio_final)
````

## Ejercicio 7 — Registro semanal de estudio ⭐⭐⭐ (20 pts)

**El programa debe pedir con `input()`:**
- La meta semanal de minutos de estudio
- Los minutos estudiados el lunes
- Los minutos estudiados el martes
- Los minutos estudiados el miércoles
- Los minutos estudiados el jueves
- Los minutos estudiados el viernes

**El programa debe imprimir con `print()`:**
```
--- RESUMEN DE ESTUDIO ---
Meta semanal: 300 minutos
Total estudiado: 255 minutos
Minutos faltantes: 45
Porcentaje de avance: 85.0 %
```


````python
# Ejercicio 7 — Registro semanal de estudio
# Escribe tu código aquí
meta = 300
print("Meta semanal: ", meta, "minutos")
total = 255
print("Total estudiados: ", total, "minutos")
minutos_faltantes = meta - total
print ("Minutos faltantes: ", minutos_faltantes)
porcentaje = total / meta * 100
print("Porcentaje de avanze:", porcentaje, "%")
````
