## Ejercicio 1 — Promedio de tiempos de entrenamiento ⭐ (10 pts)

```

```

````python
# Ejercicio 1 — Promedio de tiempos de entrenamiento
# Escribe tu código aquí
#Variables
tiempo_a = 32
tiempo_b = 28
tiempo_c = 35
#Operaciones
promedio_mins = (tiempo_a + tiempo_b + tiempo_c)/3
#Prints
print("Tiempo 1:", tiempo_a)
print("Tiempo 2:", tiempo_b)
print("Tiempo 2:", tiempo_c)
print("Promedio de minutos:", promedio_mins)
````

**Importante:** usa `sep` para separar las columnas. No necesitas usar `end`.

**Regla obligatoria:** debes usar **una sola variable de saldo** que se va actualizando paso a paso. ## Ejercicio 4 — Recargo por servicio ⭐⭐ (15 pts)

**El programa debe pedir con `input()`:**
- El precio base de una compra
- El porcentaje de recargo por servicio

**El programa debe imprimir con `print()`:**
```
--- RESUMEN DEL RECARGO ---
Precio base: $ 18000.0
Recargo (10%): $ 1800.0
Precio final: $ 19800.0
```


````python
# Ejercicio 4 — Recargo por servicio
# Escribe tu código aquí
#Variables
precio_base = float(input("Cuál es el precio base?"))
porc_recargo = int(input("Cuál es el porcentaje de recargo?"))
#Operaciones
calc_recargo = (precio_base / 100) * porc_recargo
precio_final = precio_base + calc_recargo
#Prints
print("")
print("--- RESUMEN DEL RECARGO ---")
print("Precio base:", precio_base, sep=" $ ")
print("Recargo", "(",porc_recargo,"%):", "$", calc_recargo)
print("Precio final:", precio_final, sep=" $ ")
````

## Ejercicio 5 — Repartir fondos de una actividad de curso ⭐⭐ (15 pts)

Tu curso organiza una actividad y junta dinero para financiarla.
Después de pagar los materiales, quieren saber cuánto dinero queda y cuánto corresponde por persona si se reparte en partes iguales.

**El programa debe pedir con `input()`:**
- El total reunido por el curso
- El total gastado en materiales

````python
# Ejercicio 5 — Repartir fondos de una actividad de curso
# Escribe tu código aquí
#Variables
total = float(input("Cuánto dinero hay en total?"))
gasto_mat = float(input("Cuánto dinero se gastó en materiales?"))
partic = int(input("Cuántos participantes hay?"))
#Prints
print("")
print("Total reunido:", total, sep=" $ ")
print("Gasto en materiales:", gasto_mat, sep=" $ ")
total -= gasto_mat
print("Dinero restante:", total, sep=" $ ")
print("Participantes:", partic)
total /= partic
print("Monto por persona:", total, sep=" $ ")
````

## Ejercicio 7 — Registro semanal de estudio ⭐⭐⭐ (20 pts)

**El programa debe pedir con `input()`:**
- La meta semanal de minutos de estudio
- Los minutos estudiados el lunes
- Los minutos estudiados el martes
- Los minutos estudiados el miércoles
- Los minutos estudiados el jueves
- Los minutos estudiados el viernes

**El programa debe imprimir con `print()`:**
```
--- RESUMEN DE ESTUDIO ---
Meta semanal: 300 minutos
Total estudiado: 255 minutos
Minutos faltantes: 45
Porcentaje de avance: 85.0 %
```


````python
# Ejercicio 7 — Registro semanal de estudio
# Escribe tu código aquí
#Variables
meta_semanal = int(input("Cuál es su meta semanal de minutos estudiados?"))
mins_lun = int(input("Cuántos minutos estudiados el lunes?"))
mins_mar = int(input("Cuántos minutos estudiados el martes?"))
mins_mie = int(input("Cuántos minutos estudiados el miércoles?"))
mins_jue = int(input("Cuántos minutos estudiados el jueves?"))
mins_vie = int(input("Cuántos minutos estudiados el viernes?"))
#Operaciones
total_estudiado = (mins_lun + mins_mar + mins_mie + mins_jue + mins_vie)
mins_falt = meta_semanal - total_estudiado
porcentaje_est = (total_estudiado / meta_semanal) * 100
#Prints
print("")
print("--- RESUMEN DE ESTUDIO ---")
print("Meta semanal:", meta_semanal,"minutos")
print("Total estudiado:", total_estudiado,"minutos")
print("Minutos faltantes:", mins_falt,"minutos")
print("Porcentaje de avance:", porcentaje_est,"%")

````
