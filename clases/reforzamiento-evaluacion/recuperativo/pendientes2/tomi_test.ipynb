## Ejercicio 4 — Recargo por servicio ⭐⭐ (15 pts)

**El programa debe pedir con `input()`:**
- El precio base de una compra
- El porcentaje de recargo por servicio

**El programa debe imprimir con `print()`:**
```
--- RESUMEN DEL RECARGO ---
Precio base: $ 18000.0
Recargo (10%): $ 1800.0
Precio final: $ 19800.0
```


````python

# Ejercicio 4 — Recargo por servicio
# Escribe tu código aquí
precio_base = float(input("Precio base: "))
porcentaje = float(input("Porcentaje de recargo: "))
recargo = precio_base * porcentaje / 100
precio_final = precio_base + recargo
print("  RESUMEN DEL RECARGO  ")
print("Precio base: $", precio_base)
print("Recargo (", porcentaje, "%): $", recargo)
print("Precio final: $", precio_final)
````

## Ejercicio 5 — Repartir fondos de una actividad de curso ⭐⭐ (15 pts)

Tu curso organiza una actividad y junta dinero para financiarla.
Después de pagar los materiales, quieren saber cuánto dinero queda y cuánto corresponde por persona si se reparte en partes iguales.

**El programa debe pedir con `input()`:**
- El total reunido por el curso
- El total gastado en materiales
- La cantidad de personas que participaron

**El programa debe imprimir con `print()`:**
```
--- REPARTO DE FONDOS ---
Total reunido: $ 64000.0
Gasto en materiales: $ 22000.0
Dinero restante: $ 42000.0
Participantes: 7
Monto por persona: $ 6000.0
```


````python
# Ejercicio 5 — Repartir fondos de una actividad de curso
# Escribe tu código
reunido = float(input("total reunido:"))
gastado = float(input("gastado:"))
personas = int(input("cuantas personas: "))
restante = reunido - gastado
por_persona = restante / personas
print("--- REPARTO DE FONDOS ---")
print("Total reunido: $", reunido)
print("Gasto en materiales: $", gastado)
print("Dinero restante: $", restante)
print("Participantes:", personas)
print("Monto por persona: $", por_persona)
````

## Ejercicio 6 — Depurar cálculo de promedio ponderado ⭐⭐ (15 pts)

El siguiente programa tiene **5 errores**. Encuéntralos y corrígelos sin cambiar los valores.

**El programa debe imprimir con `print()`:**
```
📋 PROMEDIO PONDERADO DEL TALLER
Ejercicios (40%): 6.0
Desafío (25%): 5.4
Bitácora (20%): 6.8
Participación (15%): 6.4
Promedio final: 6.07
```

Verificación: 6.0×0.40 + 5.4×0.25 + 6.8×0.20 + 6.4×0.15 = 6.07


````python
# Ejercicio 6 — Depurar cálculo de promedio ponderado
# Este código tiene 5 errores. Corrígelos sin cambiar los valores.

nota_ejercicios = 6.0
nota_desafio = 5.4
nota_bitacora = 6.8
nota_participacion = 6.4

promedio_final = (nota_ejercicios * 0.40) + (nota_desafio * 0.25) + (nota_bitacora * 0.20) + (nota_participacion * 0.15)

print("📋 PROMEDIO PONDERADO DEL TALLER")
print("Ejercicios (40%):", nota_ejercicios)
print("Desafío (25%):", nota_desafio)
print("Bitácora (20%):", nota_bitacora)
print("Participación (15%)", nota_participacion)
print("Promedio final:", promedio_final)
````
