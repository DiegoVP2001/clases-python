## Ejercicio 2 — Reporte de inventario del kiosco ⭐ (10 pts)

```python
producto_1 = "Jugos"
stock_1 = 24
precio_1 = 750

producto_2 = "Galletas"
stock_2 = 18
precio_2 = 600

producto_3 = "Barritas"
stock_3 = 30
precio_3 = 850

valor_total = (stock_1 * precio_1) + (stock_2 * precio_2) + (stock_3 * precio_3)

print("📦 INVENTARIO DEL KIOSCO")
print("Producto", "Stock", "Precio", sep=" | ")
print(producto_1, stock_1, "$" + str(precio_1), sep=" | ")
print(producto_2, stock_2, "$" + str(precio_2), sep=" | ")
print(producto_3, stock_3, "$" + str(precio_3), sep=" | ")
print("Valor total del inventario: $", valor_total)
```

**Pauta Ej. 2:**

- 3 pts: imprime correctamente el encabezado principal y la fila de títulos
- 3 pts: imprime el valor total del inventario con etiqueta

**Aceptable:** cualquier separador claro en `sep`, aunque no sea exactamente `" | "`. No se exige usar `end` y no se debe descontar por no usarlo.

## Ejercicio 3 — Tarjeta de transporte ⭐⭐ (15 pts)

```python
saldo_tarjeta = 12000
print("Saldo inicial: $", saldo_tarjeta)

saldo_tarjeta += 5000
print("Tras recarga mañana: $", saldo_tarjeta)

saldo_tarjeta -= 850
print("Tras viaje al liceo: $", saldo_tarjeta)

saldo_tarjeta -= 1500
print("Tras colación: $", saldo_tarjeta)

saldo_tarjeta += 3000
print("Tras recarga tarde: $", saldo_tarjeta)

saldo_tarjeta -= 850
print("Saldo final: $", saldo_tarjeta)
```

**Pauta Ej. 3:**
- 10 pts: las 5 actualizaciones son correctas, en el orden indicado, usando una sola variable de saldo
- 5 pts: imprime el saldo tras cada operación con etiqueta

**Criterio obligatorio:** debe usar la misma variable que cambia paso a paso. Es válido usar la forma corta (`+=`, `-=`) o la forma larga (`variable = variable + valor`, `variable = variable - valor`). Si crea variables separadas para cada movimiento y obtiene los resultados correctos, asignar como máximo 4/10 en cálculo, más los puntos de impresión que correspondan.

**Saldo final correcto:** $16.800

## Ejercicio 4 — Recargo por servicio ⭐⭐ (15 pts)

```python
precio_base = float(input("¿Cuál es el precio base? "))
porcentaje_recargo = float(input("¿Qué porcentaje de recargo tiene? "))

monto_recargo = precio_base * (porcentaje_recargo / 100)
precio_final = precio_base + monto_recargo

print("--- RESUMEN DEL RECARGO ---")
print("Precio base: $", precio_base)
print("Recargo (" + str(int(porcentaje_recargo)) + "%): $", monto_recargo)
print("Precio final: $", precio_final)
```

**Pauta Ej. 4:**
- 4 pts: usa `input()` para precio base y porcentaje, con conversión de tipo correcta
- 6 pts: calcula correctamente el monto del recargo y el precio final
- 5 pts: imprime el resumen con etiquetas claras

**Aceptable:** no es necesario que el porcentaje aparezca dinámico en el `print()`; se acepta `print("Recargo: $", monto_recargo)` también.

**Con 18000 y 10%:** recargo = 1800.0, precio final = 19800.0

## Ejercicio 5 — Repartir fondos de una actividad de curso ⭐⭐ (15 pts)

```python
total_reunido = float(input("¿Cuánto dinero reunió el curso? "))
gasto_materiales = float(input("¿Cuánto se gastó en materiales? "))
cantidad_participantes = int(input("¿Cuántas personas participaron? "))

dinero_restante = total_reunido - gasto_materiales
monto_por_persona = dinero_restante / cantidad_participantes

print("--- REPARTO DE FONDOS ---")
print("Total reunido: $", total_reunido)
print("Gasto en materiales: $", gasto_materiales)
print("Dinero restante: $", dinero_restante)
print("Participantes:", cantidad_participantes)
print("Monto por persona: $", monto_por_persona)
```

**Pauta Ej. 5:**
- 4 pts: usa `input()` para los 3 datos con conversión correcta (`float` para dinero, `int` para personas)
- 6 pts: calcula correctamente el dinero restante y el monto por persona
- 5 pts: imprime el resumen con las 5 líneas y etiquetas claras

**Con 64000, 22000, 7 personas:** dinero restante = 42000.0, monto por persona = 6000.0

## Ejercicio 6 — Depurar cálculo de promedio ponderado ⭐⭐ (15 pts)

**Los 5 errores eran:**
1. `nota ejercicios` → `nota_ejercicios` (espacio en nombre de variable)
2. `nota_bitacora` tenía indentación incorrecta
3. `notabitacora` → `nota_bitacora` (falta guion bajo)
4. `nota_desafio + 0.25` → `nota_desafio * 0.25` (suma en lugar de multiplicación)
5. `print("Participación (15%)" nota_participacion)` → falta la coma antes de `nota_participacion`


```python
nota_ejercicios = 6.0
nota_desafio = 5.4
nota_bitacora = 6.8
nota_participacion = 6.4

promedio_final = (nota_ejercicios * 0.40) + (nota_desafio * 0.25) + (nota_bitacora * 0.20) + (nota_participacion * 0.15)

print("📋 PROMEDIO PONDERADO DEL TALLER")
print("Ejercicios (40%):", nota_ejercicios)
print("Desafío (25%):", nota_desafio)
print("Bitácora (20%):", nota_bitacora)
print("Participación (15%):", nota_participacion)
print("Promedio final:", promedio_final)
```

## Ejercicio 7 — Registro semanal de estudio ⭐⭐⭐ (20 pts)

```python
meta_semanal = int(input("¿Cuál es tu meta semanal de estudio en minutos? "))

total_estudiado = 0
total_estudiado += int(input("Minutos del lunes: "))
total_estudiado += int(input("Minutos del martes: "))
total_estudiado += int(input("Minutos del miércoles: "))
total_estudiado += int(input("Minutos del jueves: "))
total_estudiado += int(input("Minutos del viernes: "))

minutos_faltantes = meta_semanal - total_estudiado
porcentaje_avance = total_estudiado / meta_semanal * 100

print("--- RESUMEN DE ESTUDIO ---")
print("Meta semanal:", meta_semanal, "minutos")
print("Total estudiado:", total_estudiado, "minutos")
print("Minutos faltantes:", minutos_faltantes)
print("Porcentaje de avance:", porcentaje_avance, "%")
```

**Pauta Ej. 7:**
- 5 pts: usa `input()` para la meta y los 5 días con conversión correcta
- 5 pts: acumula el total estudiado correctamente
- 5 pts: calcula minutos faltantes y porcentaje de avance correctamente
- 5 pts: imprime el resumen con las 4 líneas y etiquetas claras

**Aceptable:** guardar cada día en su propia variable y luego sumarlos.

**Con meta 300 y minutos 45+60+50+55+45:** total = 255, faltantes = 45, porcentaje = 85.0%
