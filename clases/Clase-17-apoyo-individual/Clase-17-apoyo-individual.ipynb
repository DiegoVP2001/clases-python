## 🧩 Bloque 1 — Booleanos y comparaciones

- El tipo `bool` tiene solo dos valores posibles: `True` o `False`.
- Comparadores: `==`, `!=`, `>`, `<`, `>=`, `<=`.
- Las comparaciones funcionan tanto con números como con texto.
- `bool()` convierte un número a booleano: `0` es `False`, cualquier otro número es `True`.
- ⚠️ Error típico: usar `=` (asignar) en vez de `==` (comparar).

**Ejemplo resuelto:**

```python
temperatura = 18
print("¿Hace calor?", temperatura >= 25)
>> ¿Hace calor? False

dia = "sabado"
print("¿Es fin de semana?", dia == "sabado")
>> ¿Es fin de semana? True
```

## 🧩 Bloque 2 — Operadores lógicos (`and`, `or`, `not`)

- `and`: verdadero solo si **ambas** condiciones son verdaderas.
- `or`: verdadero si **al menos una** condición es verdadera.
- `not`: invierte el valor booleano.
- Se pueden combinar con paréntesis en distinto orden: `a and (b or c)`, `(a and b) or c`, `not (a) and (b and c)`.
- ⚠️ Errores típicos: confundir `and` con `or`, escribir `true`/`false` en minúscula, olvidar paréntesis y cambiar el alcance de `not`.

**Ejemplo resuelto:**

```python
tiene_entrada = True
es_mayor_edad = False
print("¿Puede entrar solo?", tiene_entrada and es_mayor_edad)
>> ¿Puede entrar solo? False

es_socio = True
compra_sobre_10000 = False
print("¿Tiene descuento?", es_socio or compra_sobre_10000)
>> ¿Tiene descuento? True
```

## 🧩 Bloque 3 — Análisis de condiciones

- Que una condición "funcione con mis datos de prueba" **no significa que funcione siempre**.
- El **caso límite** es el valor exacto donde `>`/`>=` (o `<`/`<=`) dan resultados distintos.
- Antes de dar por buena una condición, prueba justo ese valor límite.

**Ejemplo resuelto** (con un error típico):

```python
# Una prueba exige nota mínima 4.0 para aprobar
nota = 4.0
print("¿Aprobó?", nota > 4.0)
>> ¿Aprobó? False

# Con nota exacta 4.0 debería aprobar: el operador correcto es >=
print("¿Aprobó?", nota >= 4.0)
>> ¿Aprobó? True
```

## 🧩 Bloque 4 — `if` / `else`

- Sintaxis: `if condicion:` seguido de una línea indentada; `else:` para el caso contrario.
- La indentación **es** sintaxis obligatoria en Python, no un adorno visual.
- ⚠️ Errores típicos: olvidar los `:`, no indentar el bloque, escribir un `else` sin su `if` correspondiente.

**Ejemplo resuelto:**

```python
saldo_tarjeta = 3500
precio_pasaje = 800

if saldo_tarjeta >= precio_pasaje:
    print("Puedes pagar el pasaje")
else:
    print("Saldo insuficiente")
>> Puedes pagar el pasaje
```

## 🧩 Bloque 5 — `if` anidados

- Un `if` puede ir **dentro** de otro `if`.
- La condición de adentro solo se evalúa si la condición de afuera fue `True`.
- Cada nivel de anidamiento puede tener su propio `else`.
- La sangría (indentación) marca la jerarquía: mientras más adentro, más sangría.
- Criterio: si son dos condiciones independientes que deben cumplirse ambas, muchas veces basta con `and`; anida cuando la segunda condición **depende** de que la primera se haya cumplido primero, o cuando cada rama necesita su propia lógica interna.

**Ejemplo resuelto:**

```python
edad = 20
tiene_licencia = True

if edad >= 18:
    if tiene_licencia:
        print("Puede conducir")
    else:
        print("Es mayor de edad, pero le falta la licencia")
else:
    print("Todavía no tiene edad para conducir")
>> Puede conducir
```

## 🧩 Bloque 6 — `elif`

- `elif` agrega condiciones alternativas y mutuamente excluyentes entre un `if` y un `else`.
- Solo se ejecuta **una** rama: la primera condición que resulte `True`, de arriba hacia abajo.
- El **orden importa**: las condiciones deben ir de la más específica a la más general.
- ⚠️ Errores típicos: poner una condición muy amplia primero (tapa a las siguientes), escribir un `elif` después de un `else`, olvidar el `else` final.

**Ejemplo resuelto:**

```python
puntaje = 85

if puntaje >= 90:
    print("Medalla de oro")
elif puntaje >= 70:
    print("Medalla de plata")
elif puntaje >= 50:
    print("Medalla de bronce")
else:
    print("Sin medalla")
>> Medalla de plata
```

## 🧩 Bloque 7 — Criterio de selección: ¿qué estructura usar?

- `if`/`else` simple: dos caminos, una sola condición.
- `if` anidados: una condición depende de que otra ya se haya cumplido, o cada rama necesita su propia lógica interna.
- `elif`: varias alternativas mutuamente excluyentes, ordenadas de más específica a más general.
- También se pueden **combinar**: por ejemplo, un `elif` que dentro de una de sus ramas tiene un `if`/`else` anidado.

**Ejemplo resuelto** (elif + anidado combinados):

```python
tipo_entrada = "vip"
llego_temprano = True

if tipo_entrada == "vip":
    if llego_temprano:
        print("Acceso a la zona VIP con regalo de bienvenida")
    else:
        print("Acceso a la zona VIP")
elif tipo_entrada == "general":
    print("Acceso a la zona general")
else:
    print("Entrada no válida")
>> Acceso a la zona VIP con regalo de bienvenida
```
