## 🧠 Introducción al contenido nuevo

Hasta ahora hemos usado variables para **guardar datos**.

Pero en programación, una variable también puede **cambiar su valor** durante el programa.

Ejemplo cotidiano:

> Si tienes `$10 000` en tu Cuenta RUT y recibes `$3 000`, tu saldo cambia.  
> En Python, ese cambio se puede escribir actualizando la variable.

### 1. Actualizar una variable

Observa esta línea:

```python
cuenta_rut = cuenta_rut + 3000
```

Aunque parece rara al principio, se lee así:

> Toma el valor actual de `cuenta_rut`, súmale `3000` y guarda el nuevo resultado otra vez en `cuenta_rut`.

Python primero calcula lo que está a la **derecha** del signo `=` y luego guarda el resultado en la variable de la **izquierda**.

````python
cuenta_rut = 10000

print("Antes:", cuenta_rut)

cuenta_rut = cuenta_rut + 3000

print("Después:", cuenta_rut)
````

**Pregunta rápida en + Text:**  
¿Por qué el segundo `print()` muestra un valor distinto al primero?

**¡No olvides hacer `print()` del resultado porque en caso contrario no se mostrará!**

### 2. Versión detallada y versión corta

En Python existen dos formas comunes de escribir actualizaciones.

| Versión detallada | Versión corta | Significado |
|---|---|---|
| `cuenta_rut = cuenta_rut + 1000` | `cuenta_rut += 1000` | aumenta el valor |
| `cuenta_rut = cuenta_rut - 1000` | `cuenta_rut -= 1000` | disminuye el valor |
| `cuenta_rut = cuenta_rut * 2` | `cuenta_rut *= 2` | multiplica el valor |
| `cuenta_rut = cuenta_rut / 2` | `cuenta_rut /= 2` | divide el valor |

Las dos formas sirven. La versión detallada ayuda a entender. La versión corta ayuda a escribir más limpio.

````python
# Versión detallada
cuenta_rut = 10000
cuenta_rut = cuenta_rut + 3000
print(cuenta_rut)

# Versión corta
cuenta_rut = 10000
cuenta_rut += 3000
print(cuenta_rut)
````

### 3. Otras operaciones posibles

También podemos actualizar una variable restando, multiplicando o dividiendo.

````python
cuenta_rut = 20000

cuenta_rut -= 5000   # gasto
print(cuenta_rut)

cuenta_rut *= 2      # se duplica
print(cuenta_rut)

cuenta_rut /= 4      # se divide en 4 partes
print(cuenta_rut)
````

---
## ⚠️ Errores típicos

| Error típico | Qué pasa | Cómo corregirlo |
|---|---|---|
| `cuenta_rut + 1000` | Calcula, pero no guarda el cambio | `cuenta_rut = cuenta_rut + 1000` |
| `cuenta_rut = + 1000` | Reemplaza el valor; no suma al anterior | `cuenta_rut += 1000` |
| `cuenta_rut = cuenta_rut - gasto` sin definir `gasto` | Python no sabe qué es `gasto` | definir antes: `gasto = 2500` |
| Confundir `=` con igualdad matemática | En Python `=` asigna un valor | leerlo como “guarda en” |
| Escribir `$10.000` dentro del código | Python no interpreta formato chileno con `$` y punto | usar `10000` |

> En el enunciado podemos escribir `$10 000`, pero en el código escribimos `10000`.
