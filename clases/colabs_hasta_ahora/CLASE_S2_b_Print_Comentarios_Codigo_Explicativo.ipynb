## 🧠 Introducción al contenido nuevo

Hoy veremos cómo mejorar la forma en que un programa **se explica** y **muestra resultados**.

Trabajaremos con cuatro herramientas:

| Herramienta | Para qué sirve |
|---|---|
| `#` | Escribir comentarios que Python ignora, pero que ayudan a entender el código |
| `print(texto, variable)` | Mostrar resultados con contexto |
| `sep` | Definir qué separa los elementos dentro de un `print()` |
| `end` | Definir cómo termina un `print()` |

### 1. Comentarios con `#`

Python ignora todo lo que escribas después de un `#`. Los comentarios sirven para que el código sea más fácil de entender para ti y para otras personas.

Puedes comentar de distintas formas:

<table>
<tr>
<th>Comentario general arriba de un bloque</th>
<th>Comentarios línea a línea</th>
</tr>
<tr>
<td>
<pre><code># Registro de actividades del día
canciones = 12
minutos_entrenamiento = 45
horas_estudio = 2</code></pre>
</td>
<td>
<pre><code>canciones = 12  # canciones escuchadas durante el día
minutos_entrenamiento = 45  # minutos de entrenamiento
horas_estudio = 2  # horas dedicadas a estudiar</code></pre>
</td>
</tr>
</table>

También puedes mezclar ambos estilos. Lo importante es que el comentario ayude a entender mejor el código.

> Regla simple: comenta cuando quieras explicar qué representa un bloque, un dato o un cálculo.

### 2. `print()` con texto y variables

Hasta ahora hemos usado `print()` para mostrar textos o variables por separado.

Ahora podemos mezclar texto y variables en el mismo `print()` usando comas.

````python
canciones = 12

print("Hoy escuchaste", canciones, "canciones")
````

**Resultado esperado:**

```text
Hoy escuchaste 12 canciones
```

Esto comunica mejor que mostrar solo:

```python
print(canciones)
```

### 3. Parámetro `sep`

`sep` define qué aparece entre los elementos separados por coma dentro de un `print()`.

````python
actividad = "Entrenamiento"
minutos = 45

print(actividad, minutos, "minutos", sep=" | ")
````

**Resultado esperado:**

```text
Entrenamiento | 45 | minutos
```

> `sep` no separa letras dentro de una palabra. Solo separa elementos distintos dentro del `print()`.

### 4. Parámetro `end`

`end` define cómo termina un `print()`. Por defecto, cada `print()` termina con un salto de línea.

Con `end`, podemos hacer que el siguiente `print()` continúe en la misma línea.

````python
print("Estado del resumen:", end=" ")
print("listo ✅")
````

**Resultado esperado:**

```text
Estado del resumen: listo ✅
```

## ⚠️ Errores típicos

| Error típico | Qué ocurre | Cómo corregirlo |
|---|---|---|
| `print("Puntaje" puntaje)` | Falta una coma entre texto y variable | `print("Puntaje", puntaje)` |
| `print("Puntaje" + puntaje)` | Puede dar error si `puntaje` es número | `print("Puntaje", puntaje)` |
| Creer que `sep` separa letras | `sep` solo separa elementos del `print()` | usar comas: `print("A", "B", sep="-")` |
| Usar comentarios que no ayudan a entender | El código queda más largo, pero no más claro | comentar bloques, cálculos o datos importantes |
| No usar `print()` | El cálculo ocurre, pero no se muestra | usar `print()` para mostrar el resultado |

**¡No olvides hacer `print()` del resultado porque en caso contrario no se mostrará!**

## 👥 Práctica guiada — Resumen de mis actividades del día 📊

Construiremos juntos un programa que muestre un resumen claro de algunas actividades realizadas durante el día.



### Requisitos del programa

El programa debe incluir:

1. comentarios útiles;
2. variables con nombres claros;
3. `print()` con texto y variables;
4. al menos un uso de `sep`;
5. al menos un uso de `end`.

### Resultado esperado

Tu salida debería verse así:


| Variable | Valor | Qué representa |
|---|---:|---|
