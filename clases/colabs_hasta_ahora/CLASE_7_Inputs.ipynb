## 2️⃣ Introducción al Contenido Nuevo: función `input()`

### Conceptos clave de hoy

1. `input()` detiene el programa y espera que el usuario escriba algo para personalizar la respuesta


````python
nombre_usuario = input("Ingresa tu nombre de usuario: ")

print("Tu nombre de usuario es:",nombre_usuario)
````

2. Todo lo que devuelve `input()` es **texto (string)**, aunque el usuario escriba un número


````python
edad_usuario = input("Ingresa tu edad: ")

print("Tu edad:",edad_usuario)
````

3. Si necesitamos **operar matemáticamente** los inputs,  hay que convertirlos usando:
* `int()`: para números enteros
* `float()`: para números decimales

````python
estatura_usuario = float ( input("Ingresa tu estatura en metros: ") )
edad_usuario = int( input("Ingresa tu edad en años: ") )

print( "Tu estatura multiplicada por tu edad es:", estatura_usuario*edad_usuario )
````


#### ⚠️ Errores típicos a evitar

| Error | Qué ocurre | Cómo corregirlo |
|---|---|---|
| `horas + 10` sin convertir | `TypeError` al sumar string con int | Envolver con `int()` |
| Olvidar guardar en variable | El valor se pierde | `nivel = input(...)` |
| `int()` con texto no numérico | `ValueError` | Ingresar solo números cuando se pide número |
