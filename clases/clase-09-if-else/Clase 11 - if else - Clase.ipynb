## 2️⃣ Introducción al Contenido Nuevo

Estudia cada concepto, ejecuta los ejemplos y observa el resultado:


### 💡 La sentencia `if`

ejecuta un bloque de código solo si la condición es `True`. Si la condición es `False`, el bloque se salta completamente.

**Idea clave:** _el bloque indentado solo se ejecuta si la condición es `True`._


````
saldo_bip = 3000
if saldo_bip >= 2800:
    print("Saldo suficiente. Te subes al bus.")
# >> Saldo suficiente. Te subes al bus.
````

### 💡 La cláusula `else`

define qué hacer cuando la condición del `if` es `False`. Solo uno de los dos bloques se ejecuta — nunca los dos al mismo tiempo.

**Idea clave:** _`if` y `else` son dos caminos. El programa siempre toma uno, nunca los dos._


````
saldo_bip = 1800
if saldo_bip >= 2800:
    print("Saldo suficiente. Te subes al bus.")
else:
    print("No alcanza el saldo. Se olvidó la TNE y no hay plata.")
# >> No alcanza el saldo. Se olvidó la TNE y no hay plata.
````

### 💡 La indentación

los 4 espacios (o Tab) antes de cada línea del bloque le indican a Python qué instrucciones pertenecen al `if` o al `else`. Sin indentación, Python no puede ejecutar el programa.

**Idea clave:** _la indentación no es estética, es sintaxis. Python la lee como parte del código._


````
saldo_bip = 3000
if saldo_bip >= 2800:
print("Te subes al bus.")  # sin indentación → IndentationError

# Corrección: agregar 4 espacios
if saldo_bip >= 2800:
    print("Te subes al bus.")  # funciona
````

#### ⚠️ Errores típicos a evitar

| Error | Qué ocurre | Cómo corregirlo |
|---|---|---|
| Olvidar los `:` al final del `if` o `else` | `SyntaxError` | Agregar `:` al final de la condición y después de `else` |
| No indentar el bloque | `IndentationError` | Agregar 4 espacios o Tab antes de cada línea del bloque |
| Poner `else` sin `if` previo | `SyntaxError` | `else` siempre va después de un `if` |
