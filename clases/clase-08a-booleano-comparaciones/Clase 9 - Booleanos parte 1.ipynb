### 💡 Tipo booleano

En Python existe un tipo de dato especial que solo puede tener dos valores: `True` (verdadero) o `False` (falso). Se escribe siempre con la primera letra en mayúscula. Puedes usar `type()` para confirmar que el tipo es `bool`.

**Idea clave:** _`True` y `False` son valores de datos, igual que los números._


````
tiene_stock = True
esta_agotado = False
print(tiene_stock)
print(type(tiene_stock))
````

### 💡 Operadores de comparación

Son símbolos que comparan dos valores y devuelven `True` o `False` según si la comparación se cumple o no.

  | Operador | Significado |
  |---|---|
  | `==` | Igual a |
  | `!=` | Distinto de |
  | `>` | Mayor que |
  | `<` | Menor que |
  | `>=` | Mayor o igual que |
  | `<=` | Menor o igual que |

**Idea clave:** _Una comparación siempre devuelve `True` o `False` — nunca otro valor._


````
precio_audifono = 60000
saldo_cuenta_rut = 45000
print(saldo_cuenta_rut >= precio_audifono)
````

### 🔍 Demostración: Operadores == != >

_Comparando saldo_cuenta_rut = 45000 con precio = 60000._

Ejecuta la celda y observa qué devuelve cada operador:


````
# ==
print(45000 == 60000)  # False — no son iguales

# !=
print(45000 != 60000)  # True — sí son distintos

# >
print(45000 > 60000)  # False — el saldo no supera el precio
````

### 🔍 Demostración: Operadores < >= <=

_Comparando saldo_cuenta_rut = 45000 con precio = 60000._

Ejecuta la celda y observa qué devuelve cada operador:


````
# <
print(45000 < 60000)  # True — el saldo es menor al precio

# >=
print(45000 >= 60000)  # False — el saldo no alcanza el precio

# <=
print(45000 <= 60000)  # True — el saldo está bajo o igual al precio
````

### 💡 La función `bool()`

`bool()` convierte un número en su valor booleano. El número `0` se convierte en `False`; cualquier otro número se convierte en `True`.

**Idea clave:** _`0` siempre es `False`. Cualquier otro número es `True`._


````
ventas_realizadas = 0
print(bool(ventas_realizadas))
ventas_realizadas = 150
print(bool(ventas_realizadas))
````

#### ⚠️ Errores típicos a evitar

| Error | Qué ocurre | Cómo corregirlo |
|---|---|---|
| Escribir `true` o `false` en minúscula | `NameError: name 'true' is not defined` | Siempre con mayúscula: `True`, `False` |
| Usar `=` en vez de `==` para comparar | Asigna un valor en vez de comparar | Para comparar, usar doble igual `==` |
| Comparar `"60000" >= 60000` | `TypeError` — texto vs número | Convertir con `int()` antes de comparar |
